```python
import re
import glob
```

```python
#document = 'common/buildings/00_buildings.txt'
factor = 2
```

```python
def multiplyer(path, factor):
    f = open(path, "r").read()
    f_list = list(f)
    span_list = []
    for a in re.finditer(r'(?<!(year |ince |icon |pacy |ctor |tart ))[=][ ][-]?[\d]+[.]?[\d]*',f):
        span_list.append((a.span()[0]+2,a.span()[1]))
    for i in span_list:
        number = f[i[0]:i[1]]
        f_list[i[0]] = str(float(number)*factor)
        for j in range(i[0]+1,i[1]):
            f_list[j] = ''
    f_new = ''.join(f_list)
    f = open(path, "w")
    f.write(f_new)
    f.close()
```

```python
#multiplyer(document, factor)
```

```python
common = glob.glob('common/*')
for folder in common:
    files = glob.glob(folder+'/*')
    for f in files:
        print(f)
        multiplyer(f, factor)
```
